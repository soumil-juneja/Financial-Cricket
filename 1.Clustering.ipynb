```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
#We are conducting Financial Cricket - Mock IPL auction. So for that we need fake raw data, which we generated.
```

```python
#So we created a fake raw data for batsmen, bowlers, allrounders and wicketkeeprs. Now we need to group them into 3 sets.
# marquee, set2 and set3. So, I decided to use unsupervised machine learning here.
```

```python
df = pd.read_csv("/Users/soumiljuneja/Desktop/Financial cricket/database/batsman/1_raw/player_stats_batting.csv")
```

```python
#here we have the raw data for batsmen
```

```python
df
```

```python
df.isnull().sum()
```

```python
#No missing values.
```

```python
features = ['Mat', 'Inns', 'Runs', 'HS', 'Ave', 'SR']
```

```python
df[features]
```

```python
#Standard scaled the data.
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])
X_scaled
```

```python
#We are using kmeans to cluster the data into 3 clusters.
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
df['Set'] = df['Cluster'].map({0: 'Set 2', 1: 'Marquee Set', 2: 'Set 3'}) 
```

```python
#We have succesfully clusters the data as we want.
```

```python
pd.set_option('display.max_rows', None)
df.sort_values(by = "Set")
```
