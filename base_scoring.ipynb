```python
#We min max all features to be bw 0 and 1.
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))  # Scaling between 0 and 1
features = ['Mat', 'Inns', 'Runs', 'Ave', 'SR']
X_scaled = scaler.fit_transform(df[features])
```

```python
X_scaled
```

```python
#Set different weights for different sets because they have played different no of matches.
```

```python
# Add Cluster Bonus based on the cluster column
def add_cluster_bonus(row):
    if row['Cluster'] == 1:  # Marquee set
        return 2
    elif row['Cluster'] == 0:  # Set 2
        return 1
    elif row['Cluster'] == 2:  # Set 3 (worst set)
        return 0
    return 0  # Default if no cluster or unexpected value

# Apply cluster bonus
df['Cluster Bonus'] = df.apply(add_cluster_bonus, axis=1)

# Function to calculate base score with cluster bonus and adjusted weights
def calculate_base_score_with_adjusted_weights(row):
    weights = {'Mat': 0, 'Inns': 0, 'Runs': 0, 'Ave': 0, 'SR': 0}

    # Adjust weights for players based on their cluster
    if row['Cluster'] == 1:  # Marquee set (Cluster 1)
        weights['Inns'] = 0.1
        weights['Runs'] = 0.2
        weights['Ave'] = 0.25
        weights['SR'] = 0.3
    elif row['Cluster'] == 0:  # Set 2 (Cluster 0)
        weights['Inns'] = 0.4
        weights['Runs'] = 0.45
        weights['Ave'] = 0.55
        weights['SR'] = 0.65
    elif row['Cluster'] == 2:  # Set 3 (Cluster 2)
        weights['Inns'] = 0.5
        weights['Runs'] = 0.5
        weights['Ave'] = 0.75
        weights['SR'] = 0.85

    # Compute Base Score using the scaled data and the adjusted weights
    base_score = (
        row['Mat'] * weights['Mat'] +
        row['Inns'] * weights['Inns'] +
        row['Runs'] * weights['Runs'] +
        row['Ave'] * weights['Ave'] +
        row['SR'] * weights['SR']
    )

    # Add the cluster-based bonus
    base_score += row['Cluster Bonus']

    return base_score


```

```python
df[features] = X_scaled
df
```

```python
#now, we normalise the data for the total base score of a player to be b/w 0-100.
```

```python
# Apply base score calculation
df['base_score'] = df.apply(calculate_base_score_with_adjusted_weights, axis=1)

# Normalize base score to 0-100 range based on the entire dataset
min_score = df['base_score'].min()
max_score = df['base_score'].max()

df['normalized_base_score'] = ((df['base_score'] - min_score) / (max_score - min_score)) * 100

# Final result
df
```
